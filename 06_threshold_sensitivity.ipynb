```python
# threshold_sensitivity.py
import pandas as pd

# Dados da Tabela 2 (valores reais)
models = ['Autoencoder v5', 'VQ-VAE', 'CPC', 'Predictive TF', 'Markov chain']
rratio_values = [0.570, 0.210, 0.324, 0.172, 1.000]

thresholds = [0.3, 0.5, 0.7]
results = []

for th in thresholds:
    row = [th]
    for rr in rratio_values:
        row.append(1 if rr < th else 0)   # 1 = collapse detected
    results.append(row)

df = pd.DataFrame(results, columns=['threshold'] + models)
df.to_csv('threshold_sensitivity.csv', index=False)

print("Threshold sensitivity analysis:")
print(df)
print("\nSaved threshold_sensitivity.csv")
```
